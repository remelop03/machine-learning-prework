## Introduction to Matplotlib

**Matplotlib** is a Python visualization library that provides a variety of tools and functions for creating static, animated and interactive graphics and visualizations. It is one of the most popular and widely used libraries in the Python community.

`pyplot` is a module of the Matplotlib library that provides a simple and intuitive interface for creating plots. It is typically the module used by Machine Learning and data science engineers for their graphical representations. Specifically, the key points of this module are:

- **High-level interface**: pyplot offers a number of functionalities that facilitate the fast creation of graphs.
- **Functionality**: It offers a wide variety of functions for bar charts, dot plots, box plots, and so on.
- **Integration**: It is tightly integrated with environments such as Jupyter Notebook, allowing graphs to be displayed directly within notebooks.

```python
import numpy as np

X = np.linspace(0, 10, 100)
y = np.sin(X)
z = np.cos(X)
```

### Line plot

The **plt.plot(x, y)** function takes in two inputs, a and y, and plots lines connecting pairs of (x, y) values.

```python
import matplotlib.pyplot as plt
# Data
month = ["Jan", "Feb", "Mar", "Apr", "May"]
inflation = [6.41, 6.04, 4.99, 4.93, 4.05]
# Create the line chart
fig = plt.plot(month, inflation, marker = 'o', linestyle = '-', color = 'blue')
# Customize the chart
plt.title('Inflation Trend in 2023')
plt.xlabel('Month')
plt.ylabel('Inflation')
# Display the chart
plt.tight_layout
plt.show
```

The **line plot** represents information at points connected by lines. It is useful to show the evolution of one or more data series along an axis, typically time,

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 5))

plt.plot(X, y, label = "Sen X")
plt.plot(X, z, label = "Cos X")

plt.title("Line plot")
plt.legend()
plt.show()
```

### Scatter plot

The **plt.scatter(x, y)** function takes in two inputs, x and y, and plots points representing (x, y) pairs.

```python
import matplotlib.pyplot as plt
import numpy as np
# Data
x = np.arange(1, 11, 1)
y = 2 * x
# Create the sactter plot
fig = plt.scatter(x, y, marker = 'o', color = 'blue')
# Customize the chart
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
# Display the chart
plt.tight_layout
plt.show()
```

The **scatter plot** shows individual values of two numerical variables on a Cartesian plane (with two axes). Each point represents one observation.

```python
plt.figure(figsize = (10, 5))

plt.scatter(X, y, label = "Sen X")

plt.title("Scatter plot")
plt.legend()
plt.show()
```

### Histogram

The **plt.hist(x)** function takes in one input, x, and plots a histogram of values in x to show distribution or trend.

```python
import matplotlib.pyplot as plt
import numpy as np
# Data: random 1000 samples
# data = np.random.rand(1000)
# Parameters: mean = 0, standard deviation = 1, number of values = 1000
data = np.random.normal(loc = 0, scale = 1, size = 1000)
# Create the histogram
plt.hist(data, bins = 30, edgecolor = 'black')
#Customize the chart
plt.title('Histogram of Random values')
plt.xlabel('Values')
plt.ylabel('Frequency')
# Display the chart
plt.tight_layout
plt.show() 
```

The **histogram** represents the distribution of a numerical variable by dividing the range of data into intervals and showing how many data fall into each interval (for continuous variables) or the frequency of each category (for categorical variables).

```python
data = np.random.randn(1000)

# plt.figure(figsize = (10, 5))

plt.hist(data, bins = 30, alpha = 0.7, edgecolor = 'black')

plt.title("Histogram")
plt.show()
```

### Bar chart

The **plt.bar(x, height)** funtion takes in two inputs, x and  height, and plots bars for each x value with the height given in the height variable.

```python
import matplotlib.pyplot as plt

# Data
labels = ["Course A", "Course B", "Course C"]
values = [24, 40, 30]

# Create the bar chart
fig = plt.bar(labels, values)

# Customize the chart
plt.title('Number of student in  each course')

# Display the chart
plt.show()
```

The **bar plot** represents categorical data with rectangular bars with heights (or lengths, in the case of horizontal bars) proportional to the values they represent.

```python
labels = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

plt.figure(figsize = (10, 5))

plt.bar(labels, values)

plt.title("Bar chart")
plt.show()
```

### Pie chart

A **pie chart** represents data in circular sectors, where each sector corresponds to a category and its size is proportional to the value it represents.

```python
labels = ["A", "B", "C", "D"]
sizes = [215, 130, 245, 210]

plt.figure(figsize = (7, 7))

plt.pie(sizes, labels = labels)

plt.title("Pie chart")
plt.show()
```

### Boxplot

A **boxplot** shows the distribution of quantitative data by its quartiles and possibly outliers.

![Boxplot example](https://github.com/4GeeksAcademy/machine-learning-prework/blob/main/04-matplotlib/assets/boxplot.png?raw=true "boxplot")

The ends of the box indicate the lower and upper quartiles, while the line inside the box indicates the median.

```python
data = np.random.randn(1000)

plt.figure(figsize = (10, 5))

plt.boxplot(data)

plt.title("Boxplot")
plt.show()
```
